```python
def plot_errors_on_image(image, errors):
    h, w, c = image.shape
    f, ax = plt.subplots(1, 1)
    
    
    errors = np.expand_dims(errors, 2)
    errors = np.concatenate([errors, np.zeros(shape=(h,w,c-1))], axis=2)
    
    ax.imshow(image, alpha=1)
    ax.imshow(errors, alpha=0.3)
```

```python
e = np.zeros(shape=(M.shape[:2]))
```

```python
e[50:100, 50:150]=0.5
```

```python
plot_errors_on_image(M, e)
```
