```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import os
```

```python
f, axs = plt.subplots(3,12, figsize=(12,3))
# f.subplots_adjust(0,0,-1,-1)
i = 0
for basedir, _, files in os.walk('code\sepconv-slomo\middlebury'):
    if len(files) == 1:
        continue
#     filepath = os.path.join(basedir, folder[0], files[0])
    im1, im2, im3 = (os.path.join(basedir, file) for file in files)
    
    im1 = Image.open(im1)
    im2 = Image.open(im2)
    im3 = Image.open(im3)
    
    
    axs[0,i].imshow(im1, aspect='auto')
    axs[1,i].imshow(im2,  aspect='auto')
    axs[2,i].imshow(im3, aspect='auto')
    
    axs[0,i].get_xaxis().set_visible(False)
    axs[0,i].get_yaxis().set_visible(False)
    axs[1,i].get_xaxis().set_visible(False)
    axs[1,i].get_yaxis().set_visible(False)
#     axs[2,i].get_xaxis().set_visible(False)
    axs[2,i].tick_params(axis=u'both', which=u'both',length=0)
    axs[2,i].set_xticks([])
    axs[2,i].get_yaxis().set_visible(False)
    
    axs[2,i].set(xlabel=basedir.split("\\")[-1])

    i+=1
    
plt.tight_layout()
plt.savefig('figures/middlebury.eps', bbox_inches='tight', pad_inches=0, dpi=200)
plt.savefig('figures/middlebury.png', bbox_inches='tight', pad_inches=0, dpi=200)
    
```
