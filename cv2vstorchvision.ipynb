```python
import torch
import torchvision
import cv2
import matplotlib.pyplot as plt
```

## pytorch

```python
%%time
video, _, _ = torchvision.io.read_video('redbull480.mp4')
```

```python
frame_torch = video[123].numpy()
```

## cv2

```python
%%time
cap = cv2.VideoCapture('redbull480.mp4')
cap.set(cv2.CAP_PROP_POS_FRAMES, 123)
_, frame_cv2 = cap.read()

frame_cv2 = cv2.cvtColor(frame_cv2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(frame_cv2)
```

```python
plt.imshow(frame_torch)
```

```python
(frame_cv2 == frame_torch).all()
```
